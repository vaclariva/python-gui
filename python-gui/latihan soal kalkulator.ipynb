```python
from tkinter import *
import math

# Membuat jendela utama
window = Tk()
window.title("Kalkulator GUI Dengan Python")
window.geometry('500x300')

# Mengatur latar belakang jendela
window.configure(background='lavender')

# Membuat judul "Kalkulator"
judul = Label(window, text="Kalkulator", font=("COURIER", 24), background='lavender')
judul.grid(column=1, row=0, columnspan=6)

# Membuat label dan entry untuk nilai pertama
lbl = Label(window, text="Masukkan Nilai Pertama :", anchor="e", width=25, background='lavender')
lbl.grid(column=0, row=1)
nilai1 = Entry(window, width=26)
nilai1.grid(column=1, row=1)

# Membuat label dan entry untuk nilai kedua
lbl = Label(window, text="Masukkan Nilai Kedua :", anchor="e", width=25, background='lavender')
lbl.grid(column=0, row=2)
nilai2 = Entry(window, width=26)
nilai2.grid(column=1, row=2)

# Membuat label untuk hasil perhitungan
lbl = Label(window, text="Hasil :", anchor="e", width=25, background='lavender')
lbl.grid(column=0, row=3)
hasil = Label(window, text="0", anchor="w", width=22, background='white')
hasil.grid(column=1, row=3)

# Fungsi-fungsi untuk operasi matematika
def tambah():
    hasil.configure(text=(float(nilai1.get()) + float(nilai2.get())))

def kurang():
    hasil.configure(text=(float(nilai1.get()) - float(nilai2.get())))

def kali():
    hasil.configure(text=(float(nilai1.get()) * float(nilai2.get())))

def bagi():
    if float(nilai2.get()) != 0:
        hasil.configure(text=(float(nilai1.get()) / float(nilai2.get())))
    else:
        hasil.configure(text="Error: Division by zero")

def pangkat():
    hasil.configure(text=(float(nilai1.get()) ** float(nilai2.get())))

def modulus():
    hasil.configure(text=(float(nilai1.get()) % float(nilai2.get())))

def akar():
    if float(nilai2.get()) >= 0:
        hasil.configure(text=(float(nilai1.get()) ** (1 / float(nilai2.get()))))
    else:
        hasil.configure(text="Error: Invalid input")

# Membuat tombol-tombol untuk operasi matematika
btn_tambah = Button(window, text="Tambah", command=tambah, background='white')
btn_tambah.grid(column=0, row=4)
btn_kurang = Button(window, text="Kurang", command=kurang, background='white')
btn_kurang.grid(column=1, row=4)
btn_kali = Button(window, text="Kali", command=kali, background='white')
btn_kali.grid(column=2, row=4)
btn_bagi = Button(window, text="Bagi", command=bagi, background='white')
btn_bagi.grid(column=3, row=4)
btn_pangkat = Button(window, text="Pangkat", command=pangkat, background='white')
btn_pangkat.grid(column=0, row=5)
btn_modulus = Button(window, text="Modulus", command=modulus, background='white')
btn_modulus.grid(column=1, row=5)
btn_akar = Button(window, text="Akar", command=akar, background='white')
btn_akar.grid(column=2, row=5)

# Memanggil program untuk dijalankan
window.mainloop()
```
