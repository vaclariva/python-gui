```python
from tkinter import *

# Membuat jendela utama
window = Tk()
window.title("Kalkulator GUI Dengan Python")
window.geometry('400x240')

# Membuat label untuk memasukkan nilai pertama
lbl = Label(window, text="Masukkan Nilai Pertama : ", anchor="e", width=20)
lbl.grid(column=0, row=0)

# Membuat label untuk memasukkan nilai kedua
lbl = Label(window, text="Masukkan Nilai Kedua : ", anchor="e", width=20)
lbl.grid(column=0, row=1)

# Membuat label untuk menampilkan hasil
lbl = Label(window, text="Hasil : ", anchor="e", width=20)
lbl.grid(column=0, row=2)

# Membuat entry untuk memasukkan nilai pertama
nilai1 = Entry(window, width=20)
nilai1.grid(column=1, row=0)

# Membuat entry untuk memasukkan nilai kedua
nilai2 = Entry(window, width=20)
nilai2.grid(column=1, row=1)

# Membuat label untuk menampilkan hasil perhitungan
hasil = Label(window, text="0", anchor="w", width=10)
hasil.grid(column=1, row=2)

# Fungsi-fungsi operasi matematika
def tambah():
    hasil.configure(text=(int(nilai1.get()) + int(nilai2.get())))

def kurang():
    hasil.configure(text=(int(nilai1.get()) - int(nilai2.get())))

def kali():
    hasil.configure(text=(int(nilai1.get()) * int(nilai2.get())))

def bagi():
    hasil.configure(text=(int(nilai1.get()) / int(nilai2.get())))

# Tombol-tombol operasi matematika
btn = Button(window, text="Tambah", command=tambah)
btn.grid(column=0, row=3)

btn = Button(window, text="Kurang", command=kurang)
btn.grid(column=1, row=3)

btn = Button(window, text="Kali", command=kali)
btn.grid(column=0, row=4)

btn = Button(window, text="Bagi", command=bagi)
btn.grid(column=1, row=4)

# Memulai loop utama untuk menjalankan aplikasi
window.mainloop()
```
